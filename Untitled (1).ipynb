```python
import pandas as pd

# Load the dataset
df = pd.read_csv('Sales_data.csv')

# Identify numerical columns
numerical_cols = ['Sales', 'Discounts', 'Price']

# Calculate statistics for numerical columns
for col in numerical_cols:
    mean = df[col].mean()
    median = df[col].median()
    mode = df[col].mode()[0]
    std_dev = df[col].std()
    print(f"Column: {col}")
    print(f"Mean: {mean:.2f}")
    print(f"Median: {median:.2f}")
    print(f"Mode: {mode:.2f}")
    print(f"Standard Deviation: {std_dev:.2f}")
    print()

import matplotlib.pyplot as plt

# Plot histograms for numerical columns
for col in numerical_cols:
    plt.hist(df[col], bins=10)
    plt.title(f"Histogram of {col}")
    plt.xlabel(col)
    plt.ylabel("Frequency")
    plt.show()

# Create boxplots for numerical columns
for col in numerical_cols:
    plt.boxplot(df[col])
    plt.title(f"Boxplot of {col}")
    plt.show()

# Identify categorical columns
categorical_cols = ['Category']

# Create bar charts for categorical columns
for col in categorical_cols:
    df[col].value_counts().plot(kind='bar')
    plt.title(f"Bar Chart of {col}")
    plt.xlabel(col)
    plt.ylabel("Frequency")
    plt.show()

# Standardize numerical columns
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
df[numerical_cols] = scaler.fit_transform(df[numerical_cols])

# One-hot encode categorical columns
df = pd.get_dummies(df, columns=categorical_cols)
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Load the dataset
df = pd.read_csv('Sales_data.csv')

# Print the column names
print("Column names:")
print(df.columns)

# Identify numerical columns
numerical_cols = ['Sales', 'Discounts', 'Price']

# Check if numerical columns exist in the DataFrame
for col in numerical_cols:
    if col not in df.columns:
        print(f"Error: Column '{col}' does not exist in the DataFrame.")
    else:
        mean = df[col].mean()
        median = df[col].median()
        mode = df[col].mode()[0]
        std_dev = df[col].std()
        print(f"Column: {col}")
        print(f"Mean: {mean:.2f}")
        print(f"Median: {median:.2f}")
        print(f"Mode: {mode:.2f}")
        print(f"Standard Deviation: {std_dev:.2f}")
        print()

# Plot histograms for numerical columns
for col in numerical_cols:
    if col in df.columns:
        plt.hist(df[col], bins=10)
        plt.title(f"Histogram of {col}")
        plt.xlabel(col)
        plt.ylabel("Frequency")
        plt.show()

# Create boxplots for numerical columns
for col in numerical_cols:
    if col in df.columns:
        plt.boxplot(df[col])
        plt.title(f"Boxplot of {col}")
        plt.show()

# Identify categorical columns
categorical_cols = ['Category']

# Create bar charts for categorical columns
for col in categorical_cols:
    if col in df.columns:
        df[col].value_counts().plot(kind='bar')
        plt.title(f"Bar Chart of {col}")
        plt.xlabel(col)
        plt.ylabel("Frequency")
        plt.show()

# Standardize numerical columns
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
if all(col in df.columns for col in numerical_cols):
    df[numerical_cols] = scaler.fit_transform(df[numerical_cols])

# One-hot encode categorical columns
if all(col in df.columns for col in categorical_cols):
    df = pd.get_dummies(df, columns=categorical_cols)
```
